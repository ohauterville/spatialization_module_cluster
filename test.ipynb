```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def gdp_pc_t(x):
    return 10 / (1 + np.exp(-1*(x-50)))

def m2_pc_1(x):
    return 10 / (1 + np.exp(-1*(x-10)))

def m2_pc_2(x):
    return .3 * x

def m2_pc_t1(x):
    return m2_pc_1(gdp_pc_t(x))

def m2_pc_t2(x):
    return m2_pc_2(gdp_pc_t(x))
```

```python
t = np.linspace(0, 100)
```

```python
plt.plot(t, gdp_pc_t(t))
plt.show()
```

```python
plt.plot(t, m2_pc_t1(t))
plt.plot(t, m2_pc_t2(t))
plt.show()
```
